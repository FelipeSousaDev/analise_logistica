```python
def f1(x):
    return x.max() - x.min()


# frame.apply(f1, axis='columns')


# Posso utilizar o apply para aplicar funções python aos DataFrames e Series
```

```python
def f2(x):
    return pd.Series([x.min(), x.max()], index=['min', 'max'])
```

```python
frame.apply(f2)
```

```python
#Função que entrega os valores do DataFrame formatado, ao invés de applymap, posso usar somente map

def my_format(x):
    return f"{x:.1f}"


frame.map(my_format)
```

```python
frame['e'].map(my_format)
```
