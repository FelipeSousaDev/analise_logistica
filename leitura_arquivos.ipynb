```python
datta = pd.DataFrame({"food": ["bacon", "pulled pork", "bacon", "pastrami", "corned beef", "bacon", "pastrami", "honey ham", "nova lox"],
                      "ounces": [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
datta
```

```python
meat_to_animal = {
    "bacon" : "pig",
    "pulled pork": "pig",
    "pastrami" : "cow",
    "corned beef" : "cow",
    "honey ham" : "pig",
    "nova lox" : "salmon"
}
```

```python
# O método map recebe os valores de um dicionario e cria uma nova coluna, onde a carne é correspondente ao animal, olhando para a coluna food
# e comparando com a chave + valor do dicionario meet_to_animal. 

datta['animal'] = datta['food'].map(meat_to_animal)
datta
```

```python
def get_animal(x):
    return meat_to_animal[x]
```

```python
# Usar o método map é conveniente para fazer limpezas e transformações de dados de diversos tipos. 

datta['food'].map(get_animal)
```

```python
data = pd.DataFrame(np.arange(12).reshape((3, 4)), index=['ohio', 'colorado', 'new york'], columns=['one', 'two', 'three', 'four'])
data
```

```python
# Função que transforma tanto index quanto as colunas, no caso pode ser title ou upper
def trasnform(x):
    return x[:].title()
```

```python
data.index = data.index.map(trasnform)
```

```python
data.columns = data.columns.map(trasnform)
data
```
